# Observation : when the fare is low the number of people who died is high

```python
import seaborn as sns
import matplotlib.pyplot as plt

global_mean = train_data['Survived'].mean()

for col in cat_col:
    mean_col = train_data.groupby(col)['Survived'].agg(['mean']).reset_index()
    graph = sns.barplot(x=col, y='mean', data=mean_col, palette='Greens')
    graph.axhline(global_mean, linewidth=3, color='red')
    plt.text(0, global_mean - 0.03, "global_mean", color='black', weight='semibold')
    plt.show()


```
